```python
#Counting Products with same Order IDs

dframe = data[data['Order ID'].duplicated(keep=False)]
dframe
```

```python
dframe ['Grouped'] = dframe.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
dframe.head()
```

```python
#Dropping duplicated rows

dframe1 = dframe[['Order ID', 'Grouped']].drop_duplicates()
dframe1
```

```python
#Importing libraries for counting and combinatons

from itertools import combinations
from collections import Counter
```

```python
count = Counter()

#Most Common Pairs:

for row in dframe1['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
```

```python
# Top 10 common product, which were sold together

for key, value in count.most_common(10):
    print(key, value)
```
